```python
s = np.arange(2, 21)
theory = 25 / (s * np.log(25))
evidence = np.array([2.71, 2.27, 2.17, 1.95, 1.86, 1.8, 1.61, 1.59, 1.52, 1.50, 1.44, 1.29, 1.18, 1.33, 1.08, 1.48, 1.25, 1.03, 1.50])
```

```python
plt.plot(s, theory, label='theory')
plt.plot(s, evidence, label='evidence', marker='o')
plt.xlabel('sparsity level $s$')
plt.ylabel('speedup ratio')
plt.xticks(np.arange(min(s), max(s)+1, 2.0))
plt.legend()
plt.show()
```

```python
d = np.array([10, 20, 50])
p = np.array([3, 5, 13])
theory = (d+p) / (3 * np.log(d+p))
evidence = np.array([1.70, 2.27, 2.49])

plt.plot(d, theory, label='theory')
plt.plot(d, evidence, label='evidence', marker='o')
plt.xlabel('dimensionality $d$')
plt.ylabel('speedup ratio')
plt.legend()
plt.show()
```
